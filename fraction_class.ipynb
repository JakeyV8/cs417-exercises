```python
from math import gcd
class Fraction:
  def __init__(self,top,bottom):
    self.num = top
    self.den = bottom
  def __str__(self):
    return f"{self.num}/{self.den}"
  def _reduce(self):
    comm_divisor = gcd(abs(self.num),abs(self.den))
    self.num //= comm_divisor
    self.den //= comm_divisor
    return self
  def __add__(self,other):
    new_num = self.num *other.den + self.den *other.num
    new_den = self.den*other.den
    result = Fraction(new_num, new_den)
    return result._reduce()
```

```python
f1 = Fraction(10,15)
f2 = Fraction(4,5)
result = f1.__add__(f2)
print(result)
print(f1._reduce())
```
